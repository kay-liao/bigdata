```
install.packages("ggplot2")
install.packages("SportsAnalytics")
```

```
library(SportsAnalytics)
library(ggplot2)
```

```
NBA1920 <- fetch_NBAPlayerStatistics("19-20")
```

```
str(NBA1920)
```

```
qplot(FieldGoalsAttempted, TotalPoints, data = NBA1920)
qplot(ThreesAttempted, TotalPoints, data = NBA1920)
```

```
qplot(FieldGoalsAttempted, TotalPoints, data = NBA1920, color = Position)
qplot(FieldGoalsAttempted, TotalPoints, data = NBA1920, color = Team)
qplot(ThreesAttempted, TotalPoints, data = NBA1920, color = Position)
qplot(ThreesAttempted, TotalPoints, data = NBA1920, color = Team)
```

```
qplot(FieldGoalsAttempted, TotalPoints, data = NBA1920, geom = c("point", "smooth"))
qplot(ThreesAttempted, TotalPoints, data = NBA1920, geom = c("point", "smooth"))
```

```
qplot(FieldGoalsAttempted, TotalPoints, data = NBA1920, color = Position, geom = c("point", "smooth"))
qplot(ThreesAttempted, TotalPoints, data = NBA1920, color = Position, geom = c("point", "smooth"))
```

```
qplot(TotalPoints, data = NBA1920, fill = Position)
qplot(TotalRebounds, data = NBA1920, fill = Position)
```

```
qplot(FieldGoalsAttempted, TotalPoints, 
  data = NBA1920, 
  facets = Position ~ .)
qplot(ThreesAttempted, TotalPoints, 
  data = NBA1920, 
  facets = Position ~ .)
qplot(FieldGoalsAttempted, TotalPoints, 
  data = NBA1920, 
  facets = . ~ Position)
qplot(ThreesAttempted, TotalPoints, 
  data = NBA1920, 
  facets = . ~ Position)
```

```
qplot(ThreesAttempted, TotalPoints, 
  data = NBA1920, 
  facets = Team ~ ., binwidth = 10)
qplot(ThreesAttempted, TotalPoints, 
  data = NBA1920, 
  facets = Team ~ ., binwidth = 100)
```
